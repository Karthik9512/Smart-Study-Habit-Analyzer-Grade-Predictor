```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(40)
```

```python
student= 10000
```

```python
study_hours= np.random.uniform(1,8,student)
sleep_hours= np.random.uniform(3,9,student)
social_media_hours = np.random.uniform(0,5,student)
attendance = np.random.uniform(50,100,student)
past_avg_score=np.random.uniform(50,98,student)
```

```python
student
```

```python
students= pd.DataFrame({
    'Study Hours':study_hours.round(2),
    'Sleep Hours': sleep_hours.round(2),
    'Social Media': social_media_hours.round(2),
    'Attendance': attendance.round(2),
    'Past Avg Score':past_avg_score.round(2)
    
})
```

```python
students
```

```python
final_score= (
    (study_hours*9 )+
    (sleep_hours*3)-
    (social_media_hours*2)+
    (attendance*0.3)+
    (past_avg_score*0.4)
   
)
```

```python
students['Final Score']= final_score.round(2)
```

```python
students
```

```python
students.to_csv('study_habits.csv',index= False)
```
